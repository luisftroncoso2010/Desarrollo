```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./bestsellers with categories.csv', sep=',', header=0)
df
```

```python
df.groupby('Author').count()  # Agrupamos y contamos por author
```

```python
df.groupby('Author').min()  #Buscamos la media de estos resgistro agrupados
```

```python
df.groupby('Author').max()
```

```python
df.groupby('Author').count()
```

```python
df.groupby('Author').sum().loc['William Davis']  # Filtramos los valores por loc/indices.
```

```python
# Valores planos
df.groupby('Author').sum().reset_index()
```

```python
# Agrupamiento por funciones de agregacion
df.groupby('Author').agg({'min', 'max'})
```

```python
#Agrupas por author en un diccionario ne donde solo sea dos columnas. 
df.groupby('Author').agg({'Reviews': ['min', 'max'], 'User Rating': 'sum'})
```

```python
# Agrupas por varias columnas
df.groupby(['Author', 'Year']).count()
```

```python
df.groupby('Author').agg({'Year':lambda x: [i-2000 for i in x], 'Reviews':lambda x: sum([i**2 for i in x])})
```
