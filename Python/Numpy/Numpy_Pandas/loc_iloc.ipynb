Iloc

```python
df.iloc[:]  # Muestra todo el dataFrame
```

```python
df.iloc[:, 0:3]  # Todas las filas y las columnas desde la 0 hasta la tres
```

```python
df.iloc[1, 3] * - 1  # Modificamos un valor en especifico
```
